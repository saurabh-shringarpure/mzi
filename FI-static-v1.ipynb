```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib.patches import Rectangle
import matplotlib.patches as patches
from IPython.display import HTML
from matplotlib.widgets import Slider
import ipywidgets as widgets

      
# functions that will used to interfere two waves at the second beam splitter
# interfere_amplitudes calcualtes the effective final amplitude given two waves with their amplitudes and phases
# interfere_phases calcualtes the effective final phase given two waves with their amplitudes and phases
def interfere_amplitudes(a1,phi1,a2,phi2):
    return np.sqrt(np.power((a1*np.sin(phi1)+a2*np.sin(phi2)),2)+np.power((a1*np.cos(phi1)+a2*np.cos(phi2)),2))

def interfere_phases(a1,phi1,a2,phi2):
    return np.arctan((a1*np.sin(phi1)+a2*np.sin(phi2))/(a1*np.cos(phi1)+a2*np.cos(phi2)))

def sym (lim1,lim2,si):
    return si*lim1-lim2*0**si
    
def linspace(lim1,lim2,si):
    return np.linspace(sym (lim1,lim2,si),sym(lim2,lim1,si),100)

# si = 0 creates left side of the Franson interferometer, si = 0 creates right side of the Franson interferometer
def create_figure(a,vg,c,t,ax,width1,width2,xglass,eta,k,Amp,t1,t2):
    
    
    r1 = np.sqrt(1-t1*t1); # reflectivity of the first beam splitter
    r2 = np.sqrt(1-t2*t2); # reflectivity of the first beam splitter
    
    # RIGHT
    si=1   
    width=width2
    
    x1r = linspace(0, 1,si)
    x2r = linspace(1, xglass,si)
    x3r = linspace(xglass, xglass+width,si)
    x4r = linspace(xglass+width,5,si)
    x5r = linspace(1,5,si)
    x6r = linspace(5,6,si)
    y1r = np.linspace(0,4,100)
    y2r = np.linspace(0,4,100)
    y3r = np.linspace(4,5,100)
    # preparing light waves
    rx1,=ax.plot([], [], 'r')
    rx2,=ax.plot([], [], 'r')
    rx3,=ax.plot([], [], 'b')
    rx4,=ax.plot([], [], 'r')
    rx5,=ax.plot([], [], 'r')
    rx6,=ax.plot([], [], 'r')
    ry1,=ax.plot([], [], 'r')
    ry2,=ax.plot([], [], 'r')
    ry3,=ax.plot([], [], 'r')  
    
       
    # the interferometer paths Syntax: (xdata), (ydata) for the various points
    rup = plt.Line2D((sym(1,6,si),sym(6,1,si)),(4,4),color="whitesmoke")
    rdo = plt.Line2D((sym(0,5,si),sym(5,0,si)),(0,0),color="whitesmoke")
    rle = plt.Line2D((sym(1,1,si),sym(1,1,si)),(0,4),color="whitesmoke")
    rri = plt.Line2D((sym(5,5,si),sym(5,5,si)),(0,5),color="whitesmoke")
 

    #mirrors
    rm1 = plt.Line2D((-(-1)**si*0.5,-(-1)**si*1.5),(3.5,4.5),color="black") # mirror at top left corner
    rm2 = plt.Line2D((-(-1)**si*4.5,-(-1)**si*5.5),(-0.5,0.5),color="black") # mirror at bootom right corner

    # beam spitters  (For calcualtions, first beam splitter will have reflecting surface on the left side, Seond will have reflectivng surface on the right side)
    rbs1 = plt.Line2D((-(-1)**si*0.5,-(-1)**si*1.5),(-0.5,0.5),color="grey",linestyle='dashed',lw=2.5)
    rbs2 = plt.Line2D((-(-1)**si*4.5,-(-1)**si*5.5),(3.5,4.5),color="grey",linestyle='dashed',lw=2.5)

    
    # adding elements to the figures 
    # creating the glass slide in the lower arm of the Mach-Zehnder
    rpatch = patches.Rectangle((si*xglass-(xglass+width)*0**si,-0.625), width, 1.25, facecolor='white',edgecolor='black')
    rglass=ax.add_patch(rpatch)
    ax.add_line(rup)
    ax.add_line(rdo)
    ax.add_line(rle)
    ax.add_line(rri)
    ax.add_line(rm1)
    ax.add_line(rm2)
    ax.add_line(rbs1)
    ax.add_line(rbs2)
    
    
    # creating incoming sine wave    
    rx1.set_data(x1r, Amp*np.sin(k*(x1r+(-1)**si*c*t))*np.exp(-a*(x1r+(-1)**si*vg*t)**2))

    # creating the wave transmitted through the first beam splitter    
    rx2.set_data(x2r, t1*Amp*np.sin(k*(x2r+(-1)**si*c*t))*np.exp(-a*(x2r+(-1)**si*vg*t)**2))

    # creating the wave travelling inside the glass slide    
    rx3.set_data(x3r,t1*Amp*np.sin(eta*k*(x3r+(-1)**si*xglass+(-1)**si*c*t/eta)-(-1)**si*k*xglass)*np.exp(-a*(x3r+(-1)**si*vg*t+(eta-1)*(x3r+(-1)**si*xglass))**2))

    # creating the wave emerging from the glass slide and htiing mirror m2    
    rx4.set_data(x4r,t1*Amp*np.sin(k*(x4r+(-1)**si*c*t) -(-1)**si*(eta-1)*k*width)*np.exp(-a*(x4r+(-1)**si*vg*t-(-1)**si*(eta-1)*width)**2))

    # creating the wave reflecting off the first beamsplitter (pi phase shift will occur on reflection here)    
    ry1.set_data(-(-1)**si -(-1)**si*r1*Amp*np.sin(k*(y1r-0-c*t) + np.pi + k*1)*np.exp(-a*(1+y1r-0-vg*t)**2),y1r)

    # creating the wave reflecting off the mirror m2 (pi phase shift will occur on reflection here)    
    ry2.set_data(-(-1)**si*5 -(-1)**si*t1*Amp*np.sin(k*(y2r-0-c*t) + np.pi + k*5 + (eta-1)*k*width )*np.exp(-a*(5+(eta-1)*width+y2r-0-vg*t)**2),y2r)

    # creating the wave reflecting off the mirror m1 (pi phase shift will occur on reflection here)    
    rx5.set_data(x5r,4+r1*Amp*np.sin(k*(x5r+(-1)**si*c*t)-(-1)**si* k*4)*np.exp(-a*(-(-1)**si*4+x5r+(-1)**si*vg*t)**2))

    # creating the wave emerging on the horizontal path after the second beam splitter 
    #(pi phase shift occurs for the wave in the lower path that is reflected to the right)    
    rx6.set_data(x6r,4 + Amp*interfere_amplitudes(r1*t2*np.exp(-a*(-(-1)**si*4+x6r+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*r2*np.exp(-a*(-(-1)**si*4-(-1)**si*(eta-1)*width+x6r+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width)*np.sin(k*(x6r+(-1)**si*5+(-1)**si*c*t)+interfere_phases(r1*t2*np.exp(-a*(-(-1)**si*4+x6r+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*r2*np.exp(-a*(-(-1)**si*4-(-1)**si*(eta-1)*width+x6r+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width)))

    # creating the wave emerging on the vertical path after the second beam splitter 
    #(NO phase shift occurs when the wave moving right is reflected upwards by the beam splitter. 
    # This is because we had chosen the convention that the second beam splitter has mirrored surface on the right hand side)
    ry3.set_data(-(-1)**si*5+Amp*interfere_amplitudes(r1*r2*np.exp(-a*(-(-1)**si*5-(-1)**si*y3r+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*t2*np.exp(-a*(-(-1)**si*5-(-1)**si*(eta-1)*width-(-1)**si*y3r+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width-(-1)**si*np.pi)*np.sin(k*((-1)**si*y3r+(-1)**si*4+(-1)**si*c*t)+interfere_phases(r1*r2*np.exp(-a*(-(-1)**si*5-(-1)**si*y3r+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*t2*np.exp(-a*(-(-1)**si*5-(-1)**si*(eta-1)*width-(-1)**si*y3r+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width-(-1)**si*np.pi)),y3r)
    
    
    #LEFT
    si=0  
    width=width1
    
    x1l = linspace(0, 1,si)
    x2l = linspace(1, xglass,si)
    x3l = linspace(xglass, xglass+width,si)
    x4l = linspace(xglass+width,5,si)
    x5l = linspace(1,5,si)
    x6l = linspace(5,6,si)
    y1l = np.linspace(0,4,100)
    y2l = np.linspace(0,4,100)
    y3l = np.linspace(4,5,100)
    # preparing light waves
    lx1,=ax.plot([], [], 'r')
    lx2,=ax.plot([], [], 'r')
    lx3,=ax.plot([], [], 'b')
    lx4,=ax.plot([], [], 'r')
    lx5,=ax.plot([], [], 'r')
    lx6,=ax.plot([], [], 'r')
    ly1,=ax.plot([], [], 'r')
    ly2,=ax.plot([], [], 'r')
    ly3,=ax.plot([], [], 'r')  
    
       
    # the interferometer paths Syntax: (xdata), (ydata) for the various points
    lup = plt.Line2D((sym(1,6,si),sym(6,1,si)),(4,4),color="whitesmoke")
    ldo = plt.Line2D((sym(0,5,si),sym(5,0,si)),(0,0),color="whitesmoke")
    lle = plt.Line2D((sym(1,1,si),sym(1,1,si)),(0,4),color="whitesmoke")
    lri = plt.Line2D((sym(5,5,si),sym(5,5,si)),(0,5),color="whitesmoke")
 

    #mirrors
    lm1 = plt.Line2D((-(-1)**si*0.5,-(-1)**si*1.5),(3.5,4.5),color="black") # mirror at top left corner
    lm2 = plt.Line2D((-(-1)**si*4.5,-(-1)**si*5.5),(-0.5,0.5),color="black") # mirror at bootom right corner

    # beam spitters  (For calcualtions, first beam splitter will have reflecting surface on the left side, Seond will have reflectivng surface on the right side)
    lbs1 = plt.Line2D((-(-1)**si*0.5,-(-1)**si*1.5),(-0.5,0.5),color="grey",linestyle='dashed',lw=2.5)
    lbs2 = plt.Line2D((-(-1)**si*4.5,-(-1)**si*5.5),(3.5,4.5),color="grey",linestyle='dashed',lw=2.5)

    
    # adding elements to the figures 
    # creating the glass slide in the lower arm of the Mach-Zehnder
    lpatch = patches.Rectangle((si*xglass-(xglass+width)*0**si,-0.625), width, 1.25, facecolor='white',edgecolor='black')
    lglass=ax.add_patch(lpatch)
    ax.add_line(lup)
    ax.add_line(ldo)
    ax.add_line(lle)
    ax.add_line(lri)
    ax.add_line(lm1)
    ax.add_line(lm2)
    ax.add_line(lbs1)
    ax.add_line(lbs2)
    
    # creating incoming sine wave    
    lx1.set_data(x1l, Amp*np.sin(k*(x1l+(-1)**si*c*t))*np.exp(-a*(x1l+(-1)**si*vg*t)**2))

    # creating the wave transmitted through the first beam splitter    
    lx2.set_data(x2l, t1*Amp*np.sin(k*(x2l+(-1)**si*c*t))*np.exp(-a*(x2l+(-1)**si*vg*t)**2))

    # creating the wave travelling inside the glass slide    
    lx3.set_data(x3l,t1*Amp*np.sin(eta*k*(x3l+(-1)**si*xglass+(-1)**si*c*t/eta)-(-1)**si*k*xglass)*np.exp(-a*(x3l+(-1)**si*vg*t+(eta-1)*(x3l+(-1)**si*xglass))**2))

    # creating the wave emerging from the glass slide and htiing mirror m2    
    lx4.set_data(x4l,t1*Amp*np.sin(k*(x4l+(-1)**si*c*t) -(-1)**si*(eta-1)*k*width)*np.exp(-a*(x4l+(-1)**si*vg*t-(-1)**si*(eta-1)*width)**2))

    # creating the wave reflecting off the first beamsplitter (pi phase shift will occur on reflection here)    
    ly1.set_data(-(-1)**si -(-1)**si*r1*Amp*np.sin(k*(y1l-c*t) + np.pi + k*1)*np.exp(-a*(1+y1l-vg*t)**2),y1l)

    # creating the wave reflecting off the mirror m2 (pi phase shift will occur on reflection here)    
    ly2.set_data(-(-1)**si*5 -(-1)**si*t1*Amp*np.sin(k*(y2l-c*t) + np.pi + k*5 + (eta-1)*k*width )*np.exp(-a*(5+(eta-1)*width+y2l-vg*t)**2),y2l)

    # creating the wave reflecting off the mirror m1 (pi phase shift will occur on reflection here)    
    lx5.set_data(x5l,4+r1*Amp*np.sin(k*(x5l+(-1)**si*c*t)-(-1)**si*k*4)*np.exp(-a*(-(-1)**si*4+x5l+(-1)**si*vg*t)**2))

    # creating the wave emerging on the horizontal path after the second beam splitter 
    #(pi phase shift occurs for the wave in the lower path that is reflected to the right)    
    lx6.set_data(x6l,4 + Amp*interfere_amplitudes(r1*t2*np.exp(-a*(-(-1)**si*4+x6l+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*r2*np.exp(-a*(-(-1)**si*4-(-1)**si*(eta-1)*width+x6l+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width)*np.sin(k*(x6l+(-1)**si*5+(-1)**si*c*t)+interfere_phases(r1*t2*np.exp(-a*(-(-1)**si*4+x6l+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*r2*np.exp(-a*(-(-1)**si*4-(-1)**si*(eta-1)*width+x6l+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width)))

    # creating the wave emerging on the vertical path after the second beam splitter 
    #(NO phase shift occurs when the wave moving right is reflected upwards by the beam splitter. 
    # This is because we had chosen the convention that the second beam splitter has mirrored surface on the right hand side)
    ly3.set_data(-(-1)**si*5+Amp*interfere_amplitudes(r1*r2*np.exp(-a*(-(-1)**si*5-(-1)**si*y3l+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*t2*np.exp(-a*(-(-1)**si*5-(-1)**si*(eta-1)*width-(-1)**si*y3l+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width-(-1)**si*np.pi)*np.sin(k*((-1)**si*y3l+(-1)**si*4+(-1)**si*c*t)+interfere_phases(r1*r2*np.exp(-a*(-(-1)**si*5-(-1)**si*y3l+(-1)**si*vg*t)**2),-(-1)**si*k*9,t1*t2*np.exp(-a*(-(-1)**si*5-(-1)**si*(eta-1)*width-(-1)**si*y3l+(-1)**si*vg*t)**2),-(-1)**si*k*9 -(-1)**si*(eta-1)*k*width-(-1)**si*np.pi)),y3l)
    
    
    
    #COMMON
    ax.add_patch(plt.Circle((0, 0), .125))
    
    # Axis scaling turned on so that aspect ratio is maintained
    plt.axis("scaled")
    # Axis turned off 
    plt.axis("off")
    plt.show()

style = {'description_width': 'initial'}    #to be abto to show full description of the sliders
@widgets.interact(a=widgets.FloatLogSlider(
    value=2,
    base=10,
    min=-10, # max exponent of base
    max=2, # min exponent of base
    step=0.2, # exponent step
    description='Compress wave packet',style=style
    ),vg_factor=widgets.FloatSlider(
    value=1,
    min=0,
    max=1, 
    step=0.001, 
    description='f (where group velocity = f x c)',style=style
    ),t=widgets.FloatSlider(
    value=1,
    min=0,
    max=100, 
    step=0.001, 
    description='time'
    ),c=(0,1,0.001),width1=widgets.FloatSlider(
    value=0.78,
    min=0,
    max=1, 
    step=0.001, 
    description='width of the left phase shifter',style=style
    ),width2=widgets.FloatSlider(
    value=0.78,
    min=0,
    max=1, 
    step=0.001, 
    description='width of the right phase shifter',style=style
    ),xglass=widgets.FloatSlider(
    value=2.5,
    min=1.5,
    max=3.5, 
    step=0.001, 
    description='Position of the phase shifters',style=style
    ),eta=widgets.FloatSlider(
    value=3,
    min=1,
    max=3, 
    step=0.001, 
    description='Refractive index',style=style
    ),k=(1,30,0.1),Amp=(0.062,0.25,0.001),t1=widgets.FloatSlider(
    value=0.707,
    min=0.001,
    max=0.999, 
    step=0.001, 
    description='Transmittivity of first BS',style=style
    ),t2=widgets.FloatSlider(
    value=0.707,
    min=0.001,
    max=0.999, 
    step=0.001, 
    description='Transmittivity of second BS',style=style
    ))
def update(a=0.8,vg_factor=1,c=0.2,t=0,width1 = 0.78,width2 =0.707,xglass=2.5,eta=3,k=22.2,Amp=0.125,t1=0.707,t2=0.707):
    ax=plt.gca() #clear old figure
    vg=vg_factor*c
    create_figure(a,vg,c,t,ax,width1,width2,xglass,eta,k,Amp,t1,t2)
    #select axis object contining old figure
    
```
